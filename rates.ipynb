```python
from PIL import Image, ImageTk
from tkinter import Tk, Label, Button, filedialog, PhotoImage, Canvas, StringVar
import shutil
import os
import time
import openpyxl
```

```python
class ImageRater_resized:

    def __init__(self, root):
        self.root = root
        self.root.title('rating images to help determine which propmts work')
        self.root.geometry('900x900')
        self.canvas = Canvas(root, width=800, height=800)
        self.canvas.pack()

        #self.participant = 'SN'
        #self.good_folder = f'batch/{self.participant}/good'
        #self.bad_folder = f'batch/{self.participant}/bad'

        self.participant = '011'
        self.good_folder = f'batch/{self.participant}/good'
        self.bad_folder = f'batch/{self.participant}/bad'
        
        os.makedirs(self.good_folder, exist_ok=True)
        os.makedirs(self.bad_folder, exist_ok=True)

        self.wb = openpyxl.load_workbook('interview_info.xlsx')
        self.sh = self.wb[self.participant]
        #self.sh = self.wb['test']
        self.rating_col = 'L'
        self.sets_list = []

        # column I for double prompts, K for triple
        for cell in self.sh['K']:
            if cell.value:
                self.sets_list = self.sets_list + [cell.value]
            else:
                self.sets_list = self.sets_list + ['-']
        print(self.sets_list)
        #print(self.sets_list)

        col_vals = []
        for i in range(1, self.sh.max_column + 1):
            col_vals = col_vals + [self.sh.cell(row=1, column=i).value]

        self.item_col = col_vals.index('item') + 1
        self.set_count = 0
        #self.path = f'batch/{self.participant}/'
        self.path = f'batch/{self.participant}'
        #self.image_path = f'{self.path}set_{self.set_count}/'
        self.image_path = f'{self.path}/resized'
        self.images = []
        self.current = 0

        self.label = Label(root, font=('Helvetica bold', 18), bg = 'white' , wraplength = 400, justify='center')
        self.lblVar = StringVar()
        self.progress = Label(root, font=('Helvetica bold', 14), bg = 'white' , wraplength = 400, justify='center')
        self.prgVar = StringVar()
        self.label.configure(textvariable=self.lblVar)
        self.progress.configure(textvariable=self.prgVar)

        self.root.bind('<Control-q>', lambda event: self.finish())
        self.load_img()
        
    def load_img(self):

        self.sets = [file for file in os.listdir(self.path) if 'set' in file]
        for s in self.sets:
            self.images = self.images + [f'{self.path}/{s}/{x}' for x in os.listdir(f'{self.path}/{s}') if 'png' in x]
        self.display_image()
        
    def display_image(self):
        self.root.unbind('<space>')
        if self.current < len(self.images):
            img_path = self.images[self.current]
            print(self.current)
            #self.current_set = img_path.split('/')[2]
            #self.row = self.sets_list.index(self.current_set) + 1
            self.img = PhotoImage(file=img_path, master=root)
            self.canvas.create_image(400, 400, anchor='center', image=self.img)
            #self.canvas.create_image(20, 20, anchor='center', image=self.img)
            self.prgVar.set(f'{self.current}/{len(self.images)}')
            self.progress.pack()
            x = img_path.split('/')[-1]
            set_num = x.split('_')[1]
            self.row = int(set_num) + 1
            print(f'{set_num} -- {img_path}')
            item = self.sh.cell(row=int(set_num) + 1, column=4).value
            #self.lblVar.set(img_path[16:-1])
            self.lblVar.set(f'{item} \n {set_num}')
            #print(img_path)
            self.label.pack()
            self.root.bind('<Up>', lambda event: self.good())
            self.root.bind('<Down>', lambda event: self.bad())

        else:
            #self.image_label.config(text='all images have been scored')
            self.finish()
    
    def good(self, event=None):
        #self.move_img(self.good_folder)
        if self.current < len(self.images):
            img_path = self.images[self.current]
            shutil.copy(img_path, self.good_folder)
            self.current += 1
            self.score()
            print(f'{img_path} moved to good')
            self.display_image()

    def score(self):
        if self.sh[f'M{self.row}'].value != None:
            self.sh[f'M{self.row}'].value += 1
        else:
            self.sh[f'M{self.row}'] = 1

        self.wb.save('interview_info.xlsx')
        #print(self.sh[f'M{self.row}'].value)
        
    def bad(self, event=None):
        #self.move_img(self.bad_folder)
        if self.current < len(self.images):
            img_path = self.images[self.current]
            shutil.copy(img_path, self.bad_folder)
            self.current += 1
            self.display_image()
            print(f'{img_path} moved to good')

    def finish(self):
        self.wb.save('interview_info.xlsx')
        self.wb.close()
        self.lblVar.set("all images have been scored")
        self.label.pack()
        self.canvas.delete("all")
```

```python
if __name__ == "__main__":
    root = Tk()
    rate = ImageRater_resized(root)
    root.mainloop()
```
